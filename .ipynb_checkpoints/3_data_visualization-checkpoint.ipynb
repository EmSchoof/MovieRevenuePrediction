### Distribution of Revenue in Dataframe

```python
# Import necessary modules
import scipy.stats as stats
```

```python
# Create an Empirical Cumulative Distribution Function (ECDF)
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
   
    x = np.sort(data)
    y = np.arange(1, len(data)+1) / len(data)
    return x, y
```

```python
# Seed the random number generator
np.random.seed(15)

# Define revenue
revenue = movie_generated_taglines.revenue

# Calculate ecdf in dataset
x_revenue, y_revenue = ecdf(revenue)

# Compute the theoretical CDF 
cdf_mean = np.mean(revenue)
cdf_std = np.std(revenue)

# Simulate a random sample with the same distribution and size of 10,000
cdf_samples = np.random.normal(cdf_mean, cdf_std, size=10000)

# Calculate ecdf of random sample
x_norm, y_norm = ecdf(cdf_samples)
```

```python
# Plot both ECDFs on same the figure
fig = plt.plot(x_revenue, y_revenue, marker='.', linestyle='none', alpha=0.5)
fig = plt.plot(x_norm, y_norm, marker='.', linestyle='none', alpha=0.5)

# Label figure
fig = plt.xlabel('Movie Revenue ($)')
fig = plt.ylabel('CDF')
fig = plt.legend(('Sample Population', 'Expected Norm'))
fig = plt.title('Distribution of Movie Revenue ($)')

# Save plots
plt.show()
'The sample(n=' + str(len(revenue)) + ') movie revenue $' + str(round(np.mean(revenue), 2)) + ' with a standard deviation of $' + str(round(np.std(revenue), 2)) + '.'
```

*Observations*: The overall distribution of Movie Revenue ($) is not normally distributed based on an ECDF assessment. However, under the Central Limit Theorem (CLT), this target variable can be treated as normally distributed since the sample size (n=4651) is greater than 30.

### What is the overall trend of revenue and budget over time?

```python
# Create a time subdf
time_trends = movie_df_modified[['revenue', 'budget', 'release_date_dt']]

# Create new column of the difference between budget and revenue
time_trends['difference'] = (time_trends['revenue'] - time_trends['budget'])
time_trends.head()
```

```python
# Plot Revenue over Time
time_trends.plot(x='release_date_dt', y='revenue', marker = '.', linestyle=None, figsize=(11,4))
_=plt.title('Trend of Movie Revenue Over Time')
_=plt.xlabel('Date')
_=plt.ylabel('Movie Revenue ($)')
```

```python
# Plot Budget over Time
time_trends.plot(x='release_date_dt', y='budget', marker = '.', linestyle=None, figsize=(11,4))
_=plt.title('Trend of Movie Budget Over Time')
_=plt.xlabel('Date')
_=plt.ylabel('Movie Budget ($)')
```

*Observations*: It appears that movie budgets are increasing at a faster rate than movie revenue.

```python
# Plot Net Revenue over Budget over Time
time_trends.plot(x='release_date_dt', y='difference', marker = '.', linestyle=None, figsize=(11,4))
_=plt.title('Trend of Net Movie Revenue Over Budget Over Time')
_=plt.xlabel('Date')
_=plt.ylabel('Movie Revenue Less Budget ($)')
```

### How does revenue interact with runtime columns in the dataset?

```python
# Create a time subdf
movie_runtime = movie_generated_taglines[['revenue','runtime']]
movie_runtime = movie_runtime.sort_values(by='runtime')
#movie_runtime
```

```python
# Plot Revenue by Runtime
movie_runtime.plot(x='runtime', y='revenue', kind='bar', figsize=(11,4))
_=plt.xticks(np.arange(min(movie_runtime.runtime), max(movie_runtime.runtime)+1, 10))
_=plt.title('Trend of Movie Revenue by Movie Length/Runtime')
_=plt.xlabel('Runtime (min)')
_=plt.ylabel('Movie Revenue ($)')
```
